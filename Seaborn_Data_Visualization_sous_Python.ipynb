```python
# importer la bibliothèque Seaborn et le jeu de données
import seaborn as sns
sns.set()
iris = sns.load_dataset("iris")

#Graphique linéaire
#---------------------------------------------------------------------------------
# Lineplot est le tracé le plus populaire pour représenter une relation entre deux
# caractéristiques avec la possibilité de plusieurs regroupements sémantiques.

sns.lineplot(x = iris["sepal_length"], y = iris['sepal_width'])
```

```python
# Parametre hue pour afficher un tracé pour chaque groupe
# distinct selon une troisième caractéristique

sns.lineplot(x = iris["sepal_length"], y = iris['sepal_width'],hue=iris['species'])
```

```python
# Nuage de points
#----------------------------------------------------------------------------
sns.scatterplot(x = iris["sepal_length"], y = iris['sepal_width'])
```

```python
# Bar plot, données en bandes
#---------------------------------------------------------------------------
sns.barplot(x = "petal_length",  y = "species", data=iris)
#sns.barplot(x = iris["petal_length"], y = iris['species'])
```

```python
# Pairplot
#-------------------------------------------------------------------------------
# Pour voir les relations entre les caractéristiques du jeu de données,
# on peut créer des graphiques par paire en utilisant pairplot

sns.pairplot(iris, hue="species", height = 2.5)
```

```python
# Le tracé KDE
#-------------------------------------------------------------------------------
"""
Seaborn nous fournit aussi des fonctions pour des graphiques utiles pour
l'analyse statistique. Par exemple, la fonctiondistplot permet non seulement
de visualiser l'histogramme d'un échantillon, mais aussi d'estimer la
distribution dont l'échantillon est issu.
"""
sns.distplot(iris["sepal_length"], kde=True)

# distplot sera bientot retiré de seaborn d'apres le resultat
# Penser à rempler par "displot" ou "histplot"
```

```python
# Joinplot
"""
Comme pour le pairplot que nous avons vu plus tôt, nous pouvons utiliser
jointplot pour montrer la distribution conjointe entre différentes
caractéristiques, ainsi que les distributions marginales associées
"""
sns.jointplot(x = iris["sepal_length"], y = iris['sepal_width'])
```

```python
"""
Le tracé conjoint peut même faire une estimation automatique de
la densité du noyau et une régression
"""
sns.jointplot(x = iris["sepal_length"], y = iris['sepal_width'], kind="reg")
```
