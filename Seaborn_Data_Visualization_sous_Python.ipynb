```python
# importer la bibliothque Seaborn et le jeu de donnes
import seaborn as sns
sns.set()
iris = sns.load_dataset("iris")

#Graphique linaire
#---------------------------------------------------------------------------------
# Lineplot est le trac le plus populaire pour reprsenter une relation entre deux
# caractristiques avec la possibilit de plusieurs regroupements smantiques.

sns.lineplot(x = iris["sepal_length"], y = iris['sepal_width'])
```

```python
# Parametre hue pour afficher un trac pour chaque groupe
# distinct selon une troisime caractristique

sns.lineplot(x = iris["sepal_length"], y = iris['sepal_width'],hue=iris['species'])
```

```python
# Nuage de points
#----------------------------------------------------------------------------
sns.scatterplot(x = iris["sepal_length"], y = iris['sepal_width'])
```

```python
# Bar plot, donnes en bandes
#---------------------------------------------------------------------------
sns.barplot(x = "petal_length",  y = "species", data=iris)
#sns.barplot(x = iris["petal_length"], y = iris['species'])
```

```python
# Pairplot
#-------------------------------------------------------------------------------
# Pour voir les relations entre les caractristiques du jeu de donnes,
# on peut crer des graphiques par paire en utilisant pairplot

sns.pairplot(iris, hue="species", height = 2.5)
```

```python
# Le trac KDE
#-------------------------------------------------------------------------------
"""
Seaborn nous fournit aussi des fonctions pour des graphiques utiles pour
l'analyse statistique. Par exemple, la fonctiondistplot permet non seulement
de visualiser l'histogramme d'un chantillon, mais aussi d'estimer la
distribution dont l'chantillon est issu.
"""
sns.distplot(iris["sepal_length"], kde=True)

# distplot sera bientot retir de seaborn d'apres le resultat
# Penser  rempler par "displot" ou "histplot"
```

```python
# Joinplot
"""
Comme pour le pairplot que nous avons vu plus tt, nous pouvons utiliser
jointplot pour montrer la distribution conjointe entre diffrentes
caractristiques, ainsi que les distributions marginales associes
"""
sns.jointplot(x = iris["sepal_length"], y = iris['sepal_width'])
```

```python
"""
Le trac conjoint peut mme faire une estimation automatique de
la densit du noyau et une rgression
"""
sns.jointplot(x = iris["sepal_length"], y = iris['sepal_width'], kind="reg")
```
